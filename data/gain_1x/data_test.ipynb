```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
import numpy as np
%matplotlib notebook
```

```python
c = np.loadtxt("c18-06-17-12_00_02.csv",skiprows=1,usecols=(0,1,2,3,4),delimiter=",")
print(c.shape)
```

```python
time = c[:,0]
time = (time - time[0])/3600


plt.plot(time,c[:,1],'r')
plt.plot(time,c[:,2],'g')
plt.plot(time,c[:,3],'b')
```

```python
dawn = np.where((time > 5.5) * (time < 10.))
```

```python
#print(dawn)
start = int(4.5*3600)
end = int(18.0*3600)
tbase = (c[start:end,0] - c[0,0])/3600.
#plt.plot(tbase,c[start:end,1],'r')
```

```python
red = c[start:end,1]
grn = c[start:end,2]
blu = c[start:end,3]

mmax = np.max([np.max(red), np.max(blu), np.max(grn)])

red = red/mmax
grn = grn/mmax
blu = blu/mmax




colors = []
for i, r in enumerate(red):
    colors.append(mpl.colors.to_rgb((red[i],grn[i],blu[i])))

#print(colors)
```

```python


fig, ax = plt.subplots()

#cmap = mpl.cm.cool
#colors = ["#eaa941", "#efef39", "#53a447", "#3b387f", "#48a2ba"]
cmap= mpl.colors.ListedColormap(colors)

norm = mpl.colors.Normalize(vmin=4.5, vmax=8.0)

cb1 = mpl.colorbar.ColorbarBase(ax, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')
#cb1.set_label('Some Units')
fig.show()
```
